```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('/content/HR Data.csv')
df.head()
```

```python
df.info()
```

Eliminating redundant entries.

```python
df.duplicated().sum()
```

```python
df=df.drop_duplicates()
```

Eliminate the dataset's NaN values.

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df['EmployeeCount'].unique()
```

```python
df['StandardHours'].unique()
```

```python
df['Over18'].unique()
```

 Data cleansing involves removing unnecessary columns.

```python
df=df.drop(['EmployeeCount', 'StandardHours', 'EmployeeNumber'],axis=1)
```

```python
df=df.drop(['Over18'],axis=1)
```

```python
df.columns
```

 Giving the columns new names.

```python
df=df.rename(columns={'Attrition': 'IsAttrition',
    'BusinessTravel': 'Travel_Frequency',
    'DailyRate': 'Daily_Rate_USD',
    'Education': 'Education_Level',
    'EducationField': 'Field_Of_Study',
    'TotalWorkingYears': 'Total_Experience_Years',
    'YearsAtCompany': 'Tenure_Years',
    'YearsInCurrentRole': 'Current_Role_Years',
    'YearsSinceLastPromotion': 'Years_Since_Promotion',
    'YearsWithCurrManager': 'Manager_Years'})
```

 sanitizing specific columns

```python
df= df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
df.head()
```

```python
df.skew
```

```python
x=df.drop(['IsAttrition'],axis=1)
y=df['IsAttrition']
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
le=LabelEncoder()
df['IsAttrition']=le.fit_transform(df['IsAttrition'])
df['Travel_Frequency']=le.fit_transform(df['Travel_Frequency'])
df['Field_Of_Study']=le.fit_transform(df['Field_Of_Study'])
df['Gender']=le.fit_transform(df['Gender'])
df['Department']=le.fit_transform(df['Department'])
df['JobRole']=le.fit_transform(df['JobRole'])
df['MaritalStatus']=le.fit_transform(df['MaritalStatus'])
df['OverTime']=le.fit_transform(df['OverTime'])
```

```python
x.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape,y_train.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
re=RandomForestClassifier()
re.fit(x_train,y_train)
```

```python
y_pred=re.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```
