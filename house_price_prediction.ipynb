```python
df['AREA']=df['AREA'].replace('Karapakam','Karapakkam')
df['AREA']=df['AREA'].replace('Ana Nagar','Anna Nagar')
df['AREA']=df['AREA'].replace('Ann Nagar','Anna Nagar')
df['AREA']=df['AREA'].replace('T Nagar','TNagar')
df['AREA']=df['AREA'].replace('Velchery','Velachery')
df['AREA']=df['AREA'].replace('KKNagar','KK Nagar')
df['AREA']=df['AREA'].replace('Chrompt','Chrompet')
df['AREA']=df['AREA'].replace('Chrmpet','Chrompet')
df['AREA']=df['AREA'].replace('Chormpet','Chrompet')
df['AREA']=df['AREA'].replace('Adyr','Adyar')
df['SALE_COND']=df['SALE_COND'].replace('Ab Normal','AbNormal')
df['SALE_COND']=df['SALE_COND'].replace('Partiall','Partial')
df['SALE_COND']=df['SALE_COND'].replace('PartiaLl','Partial')
df['SALE_COND']=df['SALE_COND'].replace('Adj Land','AdjLand')
df['PARK_FACIL']=df['PARK_FACIL'].replace('Noo','No')
df['BUILDTYPE']=df['BUILDTYPE'].replace('Comercial','Commercial')
df['BUILDTYPE']=df['BUILDTYPE'].replace('Other','Others')
df['UTILITY_AVAIL']=df['UTILITY_AVAIL'].replace('AllPub','All Pub')
df['UTILITY_AVAIL']=df['UTILITY_AVAIL'].replace('NoSewr ','NoSeWa')
df['STREET']=df['STREET'].replace('Pavd','Paved')
df['STREET']=df['STREET'].replace('NoAccess','No Access')
```

```python
df['N_BEDROOM']=df['N_BEDROOM'].astype(int)
df['N_BATHROOM']=df['N_BATHROOM'].astype(int)
df['QS_ROOMS']=df['QS_ROOMS'].astype(int)
df['QS_BEDROOM']=df['QS_BEDROOM'].astype(int)
df['QS_OVERALL']=df['QS_OVERALL'].astype(int)
```

```python
df.columns
df.columns

```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in df.columns:
    if df[col].dtype=='object':
        df[col]=le.fit_transform(df[col])
```

```python
plt.figure(figsize=(15,6))
sns.heatmap(df.corr())
```

```python
x=df.drop(['PRT_ID','REG_FEE','COMMIS','SALES_PRICE'],axis=1)
y=df['SALES_PRICE']
```

```python
x.shape,y.shape
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
Scaler=StandardScaler()
X_train=Scaler.fit_transform(X_train)
X_test=Scaler.fit_transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, explained_variance_score
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)
explained_variance = explained_variance_score(y_test, y_pred)


print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"R-squared (RÂ²): {r2}")
print(f"Explained Variance Score: {explained_variance}")
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor()
dtr.fit(X_train,y_train)
```

```python
dtr_pred=dtr.predict(X_test)
```

```python
dtr_mae = mean_absolute_error(y_test, dtr_pred)
dtr_mse = mean_squared_error(y_test, dtr_pred)
dtr_rmse = mean_squared_error(y_test, dtr_pred, squared=False)
dtr_r2 = r2_score(y_test, dtr_pred)
dtr_explained_variance = explained_variance_score(y_test, dtr_pred)


print(f"Mean Absolute Error (MAE): {dtr_mae}")
print(f"Mean Squared Error (MSE): {dtr_mse}")
print(f"Root Mean Squared Error (RMSE): {dtr_rmse}")
print(f"R-squared (RÂ²): {dtr_r2}")
print(f"Explained Variance Score: {dtr_explained_variance}")
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor()
rfr.fit(X_train,y_train)
```

```python
rfr_pred=rfr.predict(X_test)
```

```python
rfr_mae = mean_absolute_error(y_test, rfr_pred)
rfr_mse = mean_squared_error(y_test, rfr_pred)
rfr_rmse = mean_squared_error(y_test, rfr_pred, squared=False)
rfr_r2 = r2_score(y_test, rfr_pred)
rfr_explained_variance = explained_variance_score(y_test, rfr_pred)


print(f"Mean Absolute Error (MAE): {rfr_mae}")
print(f"Mean Squared Error (MSE): {rfr_mse}")
print(f"Root Mean Squared Error (RMSE): {rfr_rmse}")
print(f"R-squared (RÂ²): {rfr_r2}")
print(f"Explained Variance Score: {rfr_explained_variance}")
```
